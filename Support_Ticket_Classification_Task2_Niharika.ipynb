```python
# Select input and target columns

X = df['Ticket Description']

y_category = df['Ticket Type']
y_priority = df['Ticket Priority']

print("Data selected successfully!")
```

```python
# Import required libraries
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

# Download stopwords
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))

# Text cleaning function
def clean_text(text):
    text = str(text).lower()                     # lowercase
    text = re.sub(r'[^a-zA-Z\s]', '', text)     # remove punctuation/numbers
    words = text.split()                        # tokenize
    words = [word for word in words if word not in stop_words]  # remove stopwords
    return " ".join(words)

# Apply cleaning
X_cleaned = X.apply(clean_text)

# Convert text to numerical features using TF-IDF
tfidf = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf.fit_transform(X_cleaned)

print("Text cleaned and TF-IDF completed!")
print(X_tfidf.shape)
```

```python
# Split data for category classification

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,
    y_category,
    test_size=0.2,
    random_state=42
)

# Train model
category_model = MultinomialNB()
category_model.fit(X_train, y_train)

# Predict
y_pred_category = category_model.predict(X_test)

print("Ticket Category Model Trained Successfully!")
```

```python
# Evaluate category classification model

from sklearn.metrics import accuracy_score, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred_category)
print("Accuracy:", accuracy)

# Precision, Recall, F1-score
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred_category))
```

```python
# Split data for priority prediction

X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(
    X_tfidf,
    y_priority,
    test_size=0.2,
    random_state=42
)

# Train priority model
priority_model = MultinomialNB()
priority_model.fit(X_train_p, y_train_p)

# Predict priority
y_pred_priority = priority_model.predict(X_test_p)

print("Priority Prediction Model Trained Successfully!")
```

```python
# Evaluate priority prediction model

# Accuracy
priority_accuracy = accuracy_score(y_test_p, y_pred_priority)
print("Priority Prediction Accuracy:", priority_accuracy)

# Classification report
print("\nPriority Classification Report:\n")
print(classification_report(y_test_p, y_pred_priority))
```

```python
# Confusion Matrix for Priority Prediction

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test_p, y_pred_priority)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=priority_model.classes_
)

plt.figure(figsize=(8,6))
disp.plot(cmap='Blues')
plt.title("Priority Prediction Confusion Matrix")
plt.show()
```
